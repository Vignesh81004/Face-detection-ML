```python
import cv2
import os
from mtcnn import MTCNN
from google.colab.patches import cv2_imshow  # For displaying images in Colab

# User can provide a single image path or a folder path
input_path = "/content/image"  # Change this to a file path for single image processing

# Initialize MTCNN detector
detector = MTCNN()

# Check if the input path is a file or folder
if os.path.isfile(input_path):
    image_files = [input_path]  # Single image case
elif os.path.isdir(input_path):
    image_files = [os.path.join(input_path, f) for f in os.listdir(input_path)
                   if f.lower().endswith(('png', 'jpg', 'jpeg'))]  # Folder case
else:
    print("Error: Invalid path. Provide a valid image file or folder.")
    exit()

# Process each image
for img_path in image_files:
    img = cv2.imread(img_path)

    if img is None:
        print(f"Error: Could not load image {img_path}")
        continue

    detections = detector.detect_faces(img)
    total_faces = len(detections)

    print(f"\nProcessing Image: {os.path.basename(img_path)}")
    print(f"Total Faces Detected: {total_faces}")

    if total_faces == 0:
        print("No faces detected.")
        continue

    for i, det in enumerate(detections):
        x, y, w, h = det['box']
        keypoints = det['keypoints']
        confidence = det['confidence']  # Confidence score (0 to 1)

        # Determine if the face is frontal
        if 'left_eye' in keypoints and 'right_eye' in keypoints:
            left_eye_x, left_eye_y = keypoints['left_eye']
            right_eye_x, right_eye_y = keypoints['right_eye']

            eye_distance = abs(right_eye_x - left_eye_x)
            eye_alignment = abs(left_eye_y - right_eye_y)

            if eye_distance > 30 and eye_alignment < 10:
                face_accuracy = 100  # Frontal face
            else:
                face_accuracy = 75  # Side face
        else:
            face_accuracy = 75  # If one eye is missing

        # Adjust confidence to percentage and apply frontal/side accuracy logic
        final_accuracy = confidence * face_accuracy

        print(f"  Face {i+1}: Accuracy = {final_accuracy:.2f}%")

        # Draw bounding box
        cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 3)

        # Display accuracy text
        text = f"Acc: {final_accuracy:.2f}%"
        cv2.putText(img, text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Resize image for display
    scale_percent = 50  # Resize to 50% of the original size
    width = int(img.shape[1] * scale_percent / 100)
    height = int(img.shape[0] * scale_percent / 100)
    resized_img = cv2.resize(img, (width, height))

    # Show the image with face detections and accuracy
    cv2_imshow(resized_img)
```
